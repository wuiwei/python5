```python
num=int(input("请输入你要判断的正整数："))
n=num
sum=0
length=len(str(num))
#str是将参数转换成字符串的形式
#len是返回对象长度或者项目个数
#例如len('hello')=5,len([1,2,3,4,5])=5
while n!=0:
    a=n%10
    sum=sum+a**length
    n=n//10#//取整除，返回商的整数部分
if sum==num:
    print(num,"是水仙花数")
else:
    print(num,"不是水仙花数")
```
